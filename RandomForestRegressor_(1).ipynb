```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/yield_df.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
categorical = ['Area','Item']
# data['Area'] = encoder.fit_transform(data['Area'])
# data['Item'] = encoder.fit_transform(data['Item'])
for i in categorical:
  data[i] = encoder.fit_transform(data[i])
```

```python
data
```

```python
data = data.drop(['Unnamed: 0'], axis =1)
```

```python
data
```

```python
x = data.drop(['hg/ha_yield'],axis=1)
```

```python
x
```

```python
y = data['hg/ha_yield']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(xtrain),len(ytrain)
```

```python
len(xtest),len(ytest)
```

```python
# model building
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
model.predict([[0,4,1990,1485.0,121.00,16.37]])
```

```python
model.predict([[100,8,2013,657.0,2550.07,19.76]])
```

```python
# 	Area	Item	Year	average_rain_fall_mm_per_year	pesticides_tonnes	avg_temp

model.predict([[100,8,2013,657.0,2550.07,19.76]])
```

```python
# evaluation metrics : mean sqaured error, explained variance score

from sklearn.metrics import mean_squared_error, explained_variance_score
```

```python
ypred = model.predict(xtest)
```

```python
mse = mean_squared_error(ytest,ypred)
```

```python
from math import sqrt
rmse = sqrt(mse)
rmse
```

```python
evs = explained_variance_score(ytest,ypred)
evs
```
